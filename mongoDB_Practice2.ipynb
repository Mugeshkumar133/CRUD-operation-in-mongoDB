```python
ques1=[{"$group":{"_id":{"property type":"$property_type",
                         "country":"$address.country"},"max Price":{"$max":"$price",},"min price":{"$min":"$price"}}
                  },{"$sort":{"max Price":-1,"min price":-1}}]
```

```python
q1_df=list(assignment.aggregate(ques1))
```

```python
import pandas as pd
pd.DataFrame(q1_df)
```

```python
# what type of properties have more amenities        |            Ans:Apartment
```

```python
ques2=[{"$group":{"_id":{"property type":"$property_type"},"Count of amenities":{"$sum":1}}},{"$sort":{"Count of amenities":-1}}]
```

```python
q2_df=list(assignment.aggregate(ques2))
pd.DataFrame(q2_df)
```
